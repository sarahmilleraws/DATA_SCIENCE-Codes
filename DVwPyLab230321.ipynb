```python
fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(14,6))

ax[0].plot(x,y, "blue", ls="--", label=" x vs y")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
ax[0].set_xticks(x)
ax[0].set_xticklabels(x,rotation=90)
ax[0].legend()

ax[1].plot(x,z, "r", ls="-.", label="x vs z", 
           marker="+", markersize=10,markevery=5)   # 5 adımda bir + işareti koyar.
ax[1].set_xlabel("x")
ax[1].set_ylabel("z")
ax[1].legend(loc=(1.1,0.5));
```

```python

fig,ax=plt.subplots(figsize=(12,5))

data = [23, 17, 35, 29, 12, 41]
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

wp=wp={'linewidth':1,'edgecolor':'green'}
my_labels=cars
my_labels=carsmy_colors=["orange", "red", "cyan", "grey", "green", "indigo"]
my_explode=(0,0,0.2,0,0.4,0)

ax.pie(data,labels=my_labels,autopct="%1.1f%%", explode=my_explode,startangle=90,wedgeprops=wp,shadow=True)
ax.set_title("Cars")
ax.axis("equal")
ax.legend(loc=("center right"));
```

# Boxplot

```python
data=pd.read_csv("hist_data.csv")
```

```python
from scipy.stats import iqr
```

```python
data.describe().T
```

```python
data.head()
```

```python
ages=data.Age
```

```python
q75,q25=np.percentile(ages, [75,25])
```

```python
lower_bound=q25-(1.5*iqr(ages))
```

```python
upper_bound=q75+(1.5*iqr(ages))
```

```python
print("lower_band:", lower_bound)
print("upper_band:", upper_bound)
```
